# Data Viz 1

```python
# Lab 8: Data Visualization with Matplotlib
# Dataset: StudentsPerformance.csv

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (make sure the CSV is in your Colab environment)
df = pd.read_csv("/content/StudentsPerformance (1).csv")

# First look at dataset
df.head()
```

#LINE PLOT-Average subject scores


```python
avg_scores = df[['math score','reading score','writing score']].mean()

plt.figure(figsize=(8,5))
plt.plot(avg_scores, marker='o')
plt.title("Average Scores in Each Subject")
plt.ylabel("Average Score")
plt.show()

```

# Observation:
# Reading and Writing averages are slightly higher than Math.


#Histogram – Distribution of Scores

```python
plt.figure(figsize=(10,6))
plt.hist(df['math score'], bins=20, alpha=0.5, label="Math")
plt.hist(df['reading score'], bins=20, alpha=0.5, label="Reading")
plt.hist(df['writing score'], bins=20, alpha=0.5, label="Writing")
plt.title("Distribution of Math, Reading, and Writing Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.legend()
plt.show()


```

# Observation:
# Most scores are between 50–80.
# Reading and Writing are more concentrated in the higher range compared to Math.

#Pie Chart – Gender Distribution

```python
gender_counts = df['gender'].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution of Students")
plt.show()
```

# Observation:
# Gender ratio is fairly balanced with a slight difference.


#Grouped Bar Chart – Average Scores by Gender


```python
gender_scores = df.groupby('gender')[['math score','reading score','writing score']].mean()

gender_scores.plot(kind='bar', figsize=(8,6))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

# Observation:
# Females score higher in reading and writing, males slightly higher in math.

#Bar Chart – Average Scores vs Test Preparation Course

```python
prep_scores = df.groupby('test preparation course')[['math score','reading score','writing score']].mean()

prep_scores.plot(kind='bar', figsize=(8,6))
plt.title("Average Scores by Test Preparation Course")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

# Observation:
# Students who completed test preparation scored higher in all subjects,
# especially in math.

#Boxplot – Scores by Gender

```python

plt.figure(figsize=(8,6))
df.boxplot(column=['math score','reading score','writing score'], by='gender')
plt.suptitle("")
plt.title("Scores by Gender")
plt.ylabel("Score")
plt.show()
```

# Observation:
# Female students score higher in reading and writing.
# Male students have a wider spread in math scores.

#Scatter Plot – Math vs Reading

```python
plt.figure(figsize=(8,6))
plt.scatter(df['math score'], df['reading score'], alpha=0.6, c='blue')
plt.title("Math Score vs Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

# Observation:
# Positive correlation: higher math scores often mean higher reading scores.
# Some outliers exist with high reading but low math.
